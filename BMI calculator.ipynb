```python
# Body Mass Index Calculator (BMI)
from getpass import getpass

# Input name
def cal_bmi():
    # Check BMI
    print("Check your Body Mass Index.")
    print("BMI is a measurement of body fats based on height and weight\n")
    
    name = input("Enter your name: ").strip()
    print()
    
    age_ = getpass("Enter your age: ")
    print()
    
    height = float(input("Enter your height in inches: "))
    print()
    weight = float(input("Enter your weight in pounds: "))
    print("\n")

    BMI = (weight * 703) / (height * height)
    print(BMI)

    # Conditioal statement
    
    if BMI<18.5:
        print(f"{name}, you are underweight.")
    elif BMI<=24.9:
        print(f"{name}, you are normal weight.")
    elif BMI<29.9:
        print(f"{name}, you are overweight. You need to exercise more and stop sitting and writing so many python tutorials.")
    elif BMI<34.9:
        print(f"{name}, you are obese.")
    elif BMI<39.9:
        print(f"{name}, you are severely obese.")
    else:
        print(f"{name}, you are morbidly obese.")
    
cal_bmi()
```
